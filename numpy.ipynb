### Use Python's sum on Python data types and np.sum() on NumPy's ndarrays

```python
# Create massive numpy array
massive = np.random.random(10000)
massive.size
```

```python
massive[100]
```

```python
massive[:100]
```

```python
%timeit sum(massive) #time taken by Python's sum()
%timeit np.sum(massive) #time taken by NumPy's np.sum()
```

```python
np.mean(massive)
```

```python
# Standard deviation = measure of how spread out a group of numbers is from the mean
np.std(massive)
```

```python
# Variance = measure of the average degree to which each number is different from the mean
# High variance = wider range of numbers
# Low variance = lower range of numbers
np.var(massive)
```

```python
# standard deviation = sqauear root of variance
np.sqrt(np.var(massive))
```

```python
np.std(massive) == np.sqrt(np.var(massive))
```

```python
# demo of std and var
high_var_array = np.array([1,100,200,300,4000,5000])
low_var_array = np.array([2,4,6,8,10])
```

```python
(np.var(high_var_array), np.var(low_var_array))
```

```python
(np.std(high_var_array), np.std(low_var_array))
```

```python
np.mean(high_var_array), np.mean(low_var_array)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(high_var_array)
plt.plot()
```

## Dot Product Example (Nut Butter Sales)

```python
np.random.seed(0)
sales_amounts = np.random.randint(20, size=(5,3))
sales_amounts
```

```python
# Create weekly sales data frame
import pandas as pd
weekly_sales = pd.DataFrame(sales_amounts, index=["Mon", "Tue", "Wed", "Thu", "Fri"], columns=["Almond butter", "Peanut butter", "Cashew Butter"])
```

```python
weekly_sales
```

```python
# Create prices array
prices = np.array([10, 8, 12])

# Create butter_prices Data frame
butter_prices = pd.DataFrame(prices.reshape(1,3), index=["Total"], columns=["Almond butter", "Peanut butter", "Cashew Butter"])
```

```python
butter_prices
```

```python
prices.shape
```

```python
prices.T.shape
```

```python
total_sales = sales_amounts.dot(prices)
```

```python
total_sales
```

```python
total_sales.shape
```

```python
weekly_sales["Total ($)"] = total_sales
```

```python
weekly_sales
```

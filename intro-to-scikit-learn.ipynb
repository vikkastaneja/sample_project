# End to end Scikit-Learn workflow

```python
#Get the data ready
import pandas as pd
import numpy as np
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease
```

```python
# Create X - feature matrix
X = heart_disease.drop('target', axis=1)

# Create Y - labels
Y = heart_disease['target']
```

```python
# Choose the right model - since this is a classification problem, RandomForcaseClassifier is used
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)

# Choose default hyperparamters - hyperparameters are used to control/tune the learning model
clf.get_params()
```

```python
# Fit the model to training data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
clf.fit(X_train, Y_train)
```

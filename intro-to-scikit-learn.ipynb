# Option 2: fill missing values with Scikit-Learn

```python
car_missing_values = pd.read_csv('data/car-sales-extended-missing-data.csv')
car_missing_values.head()
```

```python
car_missing_values.isna().sum()
car_missing_values['Make'].isna()
```

```python
# Try without dropping the NaN price records --> DId that and when fitting, it threw an error. Hence we need to either fill the values of train and test data labels or drop the records.

car_missing_value_price_present = car_missing_values.dropna(subset='Price')
car_missing_values['Price'].isna().sum(), car_missing_value_price_present['Price'].isna().sum()
```

```python


from sklearn.model_selection import train_test_split
np.random.seed(42)
X = car_missing_value_price_present.drop('Price', axis=1)
y = car_missing_value_price_present['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# use simple imputers to fill up values as
# Make and Colour missing values will be filled in as 'missing'
# Odometer will be filled in as mean value
# Doors will be filled in by max frequency
from sklearn.impute import SimpleImputer

make_colour_imputer = SimpleImputer(strategy="constant", fill_value="missing")
odometer_imputer = SimpleImputer(strategy="mean")
door_imputer = SimpleImputer(strategy='most_frequent')

make_colour_feature = ['Make', 'Colour']
odometer_feature = ['Odometer (KM)']
door_feature = ['Doors']

# Now use transformer that will transform non-existence values to the values we defined above
from sklearn.compose import ColumnTransformer
imputer = ColumnTransformer([("make_colour_imputer", make_colour_imputer, make_colour_feature),
                             ("odometer_imputer", odometer_imputer, odometer_feature),
                             ("door_imputer", door_imputer, door_feature)])

# Calculate and fill missing values from the training set
filled_X_train = imputer.fit_transform(X_train)

# now whatever values are learned from training data, fill those into testing data
filled_X_test = imputer.transform(X_test)

# Note that we have not touched missing values for 'Price' column
Y_train.isna().count()

# filled_X_train
```
